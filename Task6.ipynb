```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Platform": ["Instagram", "Facebook", "LinkedIn", "Twitter", "YouTube"],
    "Followers": [25000, 18000, 12000, 15000, 30000],
    "Likes": [5000, 3500, 2800, 3200, 7000],
    "Comments": [800, 500, 400, 450, 1200],
    "Shares": [600, 350, 250, 300, 900]
}

df = pd.DataFrame(data)

print("========== SOCIAL MEDIA ANALYTICS REPORT ==========\n")
print(df)


df["Total_Engagement"] = (
    df["Likes"] +
    df["Comments"] +
    df["Shares"]
)

print("\n========== KPI REPORT ==========")
print("Total Followers:", df["Followers"].sum())
print("Total Likes:", df["Likes"].sum())
print("Total Comments:", df["Comments"].sum())
print("Total Shares:", df["Shares"].sum())
print("Total Engagement:", df["Total_Engagement"].sum())


best_platform = df.loc[df["Total_Engagement"].idxmax()]
print("\nBest Performing Platform")
print(best_platform)


print("\n========== SUMMARY ==========")
print(df.describe())

plt.figure(figsize=(8,5))
plt.bar(df["Platform"], df["Followers"])
plt.title("Followers by Platform")
plt.xlabel("Platform")
plt.ylabel("Followers")
plt.show()


plt.figure(figsize=(8,5))
plt.plot(df["Platform"], df["Total_Engagement"], marker='o')
plt.title("Engagement Analysis")
plt.xlabel("Platform")
plt.ylabel("Engagement")
```
